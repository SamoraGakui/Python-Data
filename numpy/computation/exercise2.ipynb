How to add a border (filled with 0's) around an existing array?

```python
Z = np.ones((5,5))
Z = np.pad(Z, pad_width=1, mode='constant', constant_values=0)
print(Z)
```

Create a 8x8 matrix and fill it with a checkerboard pattern

```python
x = np.zeros((8, 8))
# Set 1s on alternating columns starting from index 1 for even rows
x[::2, 1::2] = 1

# Set 1s on alternating columns starting from index 0 for odd rows
x[1::2, ::2] = 1

x
```

Create a checkerboard 8x8 matrix using the tile function

```python
x = np.tile( np.array([[0,1],[1,0]]), (4,4))
print(x)
```

Normalize a 5x5 random matrix

```python
Z = np.random.random((5,5))
Z = (Z - np.mean (Z)) / (np.std (Z))
print(Z)
```

Create a custom dtype that describes a color as four unsigned bytes (RGBA)

```python
color = np.dtype([("r", np.ubyte, 1),
                  ("g", np.ubyte, 1),
                  ("b", np.ubyte, 1),
                  ("a", np.ubyte, 1)])
color
```

Multiply a 5x3 matrix by a 3x2 matrix (real matrix product)

```python
X = np.ones((5,3)) @ np.ones((3,2))
print(X)
```

Given a 1D array, negate all elements which are between 3 and 8, in place.

```python
Z = np.arange(11)
Z[(3 < Z) & (Z <= 8)] *= -1
print(Z)
```

How to round away from zero a float array ?

```python
x = np.random.uniform(-10,+10,10)
print (np.copysign(np.ceil(np.abs(x)), x))
```

How to find common values between two arrays?

```python
x1 = np.random.randint(0,10,10)
print(x1)
x2 = np.random.randint(0,10,10)
print(x2)
print(np.intersect1d(x1,x2))
```
