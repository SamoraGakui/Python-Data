```python
import numpy as np
np.__version__
```

```python
np.show_config()
```

```python
# Create a null vector of size 10
np.zeros(10)
```

```python
# itemsize, which lists the size (in bytes) of each array element,
x = np.zeros(10)
print("itemsize:", x.itemsize, "bytes")
print("Size:", x.size, "bytes")
# nbytes, which lists the total size (in bytes) of the array
print("nbytes:", x.nbytes, "bytes")
```

```python
# Create a null vector of size 10 but the fifth value which is 1
x[4] = 5
print(x)
```

```python
# Create a vector with values ranging from 10 to 49
y = np.arange(10, 50)
print(y)
```

```python
# Reverse a vector (first element becomes last)
y[::-1]
```

```python
# Create a 3x3 matrix with values ranging from 0 to 8
grid = np.arange(0, 9).reshape((3, 3))
print(grid)
```

```python
# Find indices of non-zero elements from [1,2,0,0,4,0]
np.nonzero([1,2,0,0,4,0])
```

```python
# Create a 3x3 identity matrix
np.eye(3)
```

```python
# Create a 3x3x3 array with random values
np.random.randint(0, 10, (3, 3))
```

```python
np.random.random((3, 3, 3))
```

```python
# Create a 10x10 array with random values and find the minimum and maximum values
grid = np.random.random((10, 10))
min = grid.min()
max = grid.max()
print(min, max)
```

```python
# Create a random vector of size 30 and find the mean value
np.random.random(30).mean()
```

```python
# 1. Create a 2D array filled entirely with 1s
arr = np.ones((5, 5), dtype=int)

# 2. Slice the inner elements and set them to 0
arr[1:-1, 1:-1] = 0

print(arr)
```
