### Step 9. Which was the most-ordered item?

```python
# 1. Remove the dollar sign and strip any blank spaces
chipo["item_price"] = chipo["item_price"].str.replace("$", "", regex=False).str.strip()

# 2. Convert to float first (required for decimal numbers)
chipo["item_price"] = chipo["item_price"].astype(float)
```

```python
chipo.groupby('item_name')['item_price'].sum().sort_values(ascending=False)
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
# Group by the item name and sum the quantities
item_totals = chipo.groupby("item_name")["quantity"].sum()

# Find the item with the highest total
most_ordered_item = item_totals.idxmax()
highest_quantity = item_totals.max()

print(f"Most ordered item: {most_ordered_item}")
print(f"Total items ordered: {highest_quantity}")
```

### Step 11. What was the most ordered item in the choice_description column?

```python
chipo.groupby(['item_name','choice_description'])['quantity'].sum()
```

### Step 12. How many items were orderd in total?

```python
# Sum up the entire quantity column
total_items = chipo["quantity"].sum()

print(f"Total items ordered: {total_items}")
```
